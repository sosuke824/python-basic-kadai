```python
last_name = "侍"
first_name = "太郎"
sister_name = "花子"

# 3つの変数を文字列内に埋め込んで表示
print(f"私の名前は{last_name}{first_name}です。妹の名前は{sister_name}です。")
```

```python
# 変数の設定
jouhen = 10
kahen = 20
takasa = 5
# 台形の面積の計算
menseki = (jouhen + kahen) * takasa / 2
# 単位の設定
unit1 = "cm"
unit2 = "㎠"
# 各値の表示
print(f"上辺：{jouhen}{unit1}")
print(f"下辺：{kahen}{unit1}")
print(f"高さ：{takasa}{unit1}")
# 台形の面積の計算結果
print(f"台形の面積 = {menseki}{unit2}")
```
