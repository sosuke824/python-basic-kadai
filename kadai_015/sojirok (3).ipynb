```python
#ランダムな整数を利用するために、randomモジュールをインポート
import random

# 変数numに0～4までのランダムな整数を代入する
num = random.randint(0, 4)

# 変数numの値を出力
print(num)

# すべての条件が成り立つ場合にのみ処理を行う（変数numの値が2)
if 1 < num and num < 3:
  print("変数numは1より大きく、3より小さいです")
else:
  print("and条件が成り立ちませんでした")

# 1つでも条件が成り立てば処理を行う（変数numの値が1または3）
if num == 1 or num == 3:
  print("変数numは1または3です")
else:
  print("or条件が成り立ちませんでした")
```

```python
#ランダムな整数を利用するために、randomモジュールをインポート
import random

# 変数varに1～20までのランダムな整数を代入する
var = random.randint(1, 20)

# 変数varが、3の倍数の場合「Fizz」を出力、5の倍数の場合「Buzz」を出力、3の倍数と5の倍数の両方に該当する場合は「FizzBuzz」を出力、どの場合にも該当しない場合は変数varの値を出力
if var % 3 == 0 and var % 5 != 0:
  print("Fizz")
elif var % 5 == 0 and var % 3 != 0:
  print("Buzz")
elif var % 3 == 0 and var % 5 == 0:
  print("FizzBuzz")
else:
  print(var)
```

```python
import random
num = random.randint(0,4)

# カウンタ変数を用意する
i = 1

# 変数numの最初の値を出力する（確認用）
print(f"最初の値は{num}です。")

# 変数numの値が以外である間、変数numの値を出力し続ける
while num != 0:
  # 変数numに0~4までのランダムな整数を代入する
  num = random.randint(0,4)

  # カウンタ変数iの値が5であれば、break文で繰り返し処理を強制終了する
  if i == 5:
    print("5回目なので繰り返し処理を強制終了します。")
    break

  # 次の条件式で比較される、変数numの現在の値を出力する
  print(f"現在の値は{num}です。")
  i = i + 1
```

```python
#ランダムな整数を利用するため、記述が必要です。
import random

# 合計用の変数を用意する
sum = 0

# 変数sumの値が20以上になるまで繰り返し処理を行う
while sum < 20:
    # 変数numに1~10までのランダムな整数を代入する
    num = random.randint(1, 10)

    print(f"{num}が出ました。")

    # 変数numの値が偶数（2で割った余りが0）であれば、加算せずにcontinue文で次のループに進む
    if num % 2 == 0:
        print("偶数なので加算しません。")
        continue

    # 変数sumに変数numの値を加算する
    sum = sum + num
    print(f"現在の合計は{sum}です。")
```

```python
from datetime import datetime

class Event:
    def __init__(self, name: str, date: str, location: str):
        self.name = name
        self.date = date
        self.location = location

    def days_until_event(self) -> int:
        today = datetime.today()
        diff = self.date - today
        return diff.days


# ===== インスタンス作成（未来の日付）=====
event_date = datetime(2026, 4, 1)  # 未来の日付に設定

event = Event(
    "プログラミング初心者向けセミナー",
    event_date,
    "東京国際フォーラム"
)

# ===== 属性の出力 =====
print("イベント名:", event.name)
print("開催日:", event.date)
print("開催場所:", event.location)

# ===== 開催日までの日数 =====
print("開催日までの日数:", event.days_until_event(), "日")
```

```python
class Human:
    # コンストラクタ
    def __init__(self, name: str, age: int):
        self.name = name
        self.age = age

    # 情報を出力するメソッド
    def printinfo(self):
        print(self.name)
        print(self.age)

# インスタンスを作成
human = Human("山田太郎", 25)

# メソッドを呼び出して情報を表示
human.printinfo()
```
